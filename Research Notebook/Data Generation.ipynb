```python
import pandas as pd
```

```python
df_videos = pd.read_csv("../data/SharedSkillet.csv", index_col=False)
df_text = pd.read_csv("../data/Recipe.csv", index_col=False)
```

```python
df_videos.head()
```

```python
df_text.head()
```

```python
# Clean column names by stripping extra whitespace and tab characters
df_videos.columns = df_videos.columns.str.strip()
df_text.columns = df_text.columns.str.strip()

# Clean the "File Name" column by stripping extra whitespace and tab characters
df_videos["File Name"] = df_videos["File Name"].str.strip()
df_text["File Name"] = df_text["File Name"].str.strip()
```

```python
df_text["File Name"]
```

```python

# Perform an outer join to retain all records
merged_df = pd.merge(df_videos, df_text, on="File Name", how="outer")

# Remove duplicate rows based only on the "File Name" column
merged_df = merged_df.drop_duplicates(subset=["File Name"])
```

```python
merged_df['Title'].value_counts()
```

```python
merged_df.drop('Title', inplace = True, axis = 1)
```

```python
merged_df['Title'] = merged_df['File Name'].apply(lambda x: x.rsplit('.', 1)[0])
```

```python
merged_df.head()
```

```python
# Save the cleaned and merged file
merged_df.to_csv("merged_output.csv", index=False)

print("Merging complete. Cleaned file saved as 'merged_output.csv'.")
```

```python
# Import pandas and display module
import pandas as pd
from IPython.display import HTML, display

# Read the CSV file (adjust path to your Complete.csv location)
try:
    # Update this path to where your CSV is (e.g., "Complete.csv" or "C:/Users/YourName/Complete.csv")
    df = pd.read_csv("..\data\Complete.csv", index_col=False, encoding='unicode_escape' )  # Change this!
    html = ""

    # Generate HTML for each recipe
    for index, row in df.iterrows():
        # Extract video_id from Public URL if it exists
        public_url = row.get("Public URL", "")
        video_id = public_url.split("/d/")[1].split("/")[0] if pd.notna(public_url) and "drive.google.com" in public_url else ""

        # Get title and description
        title = row.get("Title", row.get("title", ""))
        description = row.get("Recipe in Plain text", row.get("description", ""))  # Prefer Recipe in Plain text

        # Validate and process
        if video_id and title:
            html += '<div class="recipe-card">\n'
            html += f'    <iframe src="https://drive.google.com/file/d/{video_id}/preview" allow="autoplay"></iframe>\n'
            html += f'    <h3>{title}</h3>\n'
            html += f'    <p>{description if pd.notna(description) else "No description available"}</p>\n'
            html += '</div>\n'
        else:
            print(f"Skipping row {index} due to missing video_id or title. Public URL: {public_url}, Title: {title}")

    # Display the HTML in the notebook
    display(HTML(html))

    # Save to a file
    with open("recipe_cards.html", "w", encoding="utf-8") as f:
        f.write(html)
    print("HTML saved to recipe_cards.html")

except FileNotFoundError:
    print("Error: Complete.csv not found. Check the file path.")
except KeyError as e:
    print(f"Error: Column {e} not found in CSV. Check your column names.")
except Exception as e:
    print(f"An error occurred: {e}")
```
